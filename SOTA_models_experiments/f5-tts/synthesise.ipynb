```python
import os
import f5_tts
from f5_tts.api import F5TTS

# Bundled English example clip shipped with the f5-tts package (handy smoke-test default).
# Transcript: "Some call me nature, others call me mother nature."
_F5_PKG = list(f5_tts.__path__)[0]
DEFAULT_REF_EN = os.path.join(_F5_PKG, 'infer/examples/basic/basic_ref_en.wav')


def synthesise_english(
    sentences,
    output_dir='outputs/f5tts',
    label=None,
    model='F5TTS_v1_Base',
    ref_file=DEFAULT_REF_EN,
    ref_text='Some call me nature, others call me mother nature.',
    seed=None,
    f5tts=None,
):
    """Synthesise a list of English sentences with F5-TTS via zero-shot voice cloning.

    Mirrors CosyVoice/synthesise.ipynb::synthesise_english so the two models can be
    compared under an identical zero-shot reference-cloning protocol.

    Args:
        sentences:  list[str] of English sentences to synthesise.
        output_dir: where the .wav files are written.
        model:      F5-TTS model id (downloaded from HuggingFace on first use).
        ref_file:   reference clip whose voice/accent is cloned. Defaults to the
                    package's English example; for an accented-TTS study, swap this
                    for an English accented reference clip.
        ref_text:   transcript of `ref_file` (must match the audio).
        seed:       optional int for reproducible generation.
        f5tts:      a pre-loaded F5TTS to reuse; if None, one is loaded.

    Returns:
        (f5tts, paths): the loaded model (for reuse) and list of written wav paths.
    """
    os.makedirs(output_dir, exist_ok=True)
    # Expand ~ so home-relative reference paths resolve (open() does not expand ~).
    ref_file = os.path.expanduser(ref_file)
    if f5tts is None:
        f5tts = F5TTS(model=model)

    paths = []
    for idx, text in enumerate(sentences):
        out_path = os.path.join(output_dir, 'sentence_{}_{:02d}.wav'.format(label,idx))
        f5tts.infer(
            ref_file=ref_file,
            ref_text=ref_text,
            gen_text=text,
            file_wave=out_path,
            seed=seed,
        )
        paths.append(out_path)
        print('saved', out_path)
    return f5tts, paths
```

```python
english_sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "She sells seashells by the seashore on a sunny afternoon.",
    "Please call Stella and ask her to bring these things from the store.",
    "I really think we should leave before the weather gets any worse.",
    "Authentication failed, so the researcher tried a different approach.",
]
```

```python
f5tts, paths = synthesise_english(english_sentences)
paths
```
